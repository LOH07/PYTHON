```python
# the purpose of this code is to:
# 1) find all employees whose first name is "John"
# 2) select those of them whose job is "Stage Electrician"
# 3) select those of them whose BasePay is greater than 1000
# that requires 3 conditions


# import the necessary library
import pandas as pd


# read data from CSV file
df = pd.read_csv("Salaries.csv")


# apply the 3 conditions for selection
df[(df["EmployeeName"].apply(lambda x: x[0:4]=="John")) & (df["JobTitle"]=="Stage Electrician") & (df["BasePay"]>1000)]
```
