```python
# Import the necessary libraries: Pandas and Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read from CSV file
df = pd.read_csv("Salaries.csv")

# Create a Dictionary of the 10 most common Jobs in the CSV file
d = dict(df["JobTitle"].value_counts().head(10))

# Create a list of values for Explode parameter of the Matplotlib PIE
exp = [(x==max(d.values()))*0.1 for x in d.values()]

# Create the Matplotlib PIE as a Subplot
fig,axes = plt.subplots(nrows=1,ncols=1,dpi=100)
axes.pie(d.values(),explode=exp,labels=d.keys(),autopct="%1.1f%%",startangle=90,pctdistance=0.8)

# Display the Matplotlib PIE
plt.show()
```
